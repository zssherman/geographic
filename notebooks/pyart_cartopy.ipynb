## Py-ART Cartopy Plotting

Testing out pyarts new cartopy function.

```python
#Reading in a NEXRAD file and plotting a PPI.
#Import modules
import os
import pyart
import numpy as np
import matplotlib.pyplot as plt
import scipy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
%matplotlib inline
print(pyart.__version__)
```

```python
# Importing datafile
filename = '/home/zsherman/CSU_unfolded/CSU_unfolded/CPOL_corr_20060122_123009_vt_unfolded_hand.cfrad.nc'
radar = pyart.io.read(filename)
```

```python
# Simple cartopy ppi.
fig = plt.figure(figsize=(12, 8))

display = pyart.graph.RadarMapDisplayCartopy(radar)
display.plot_ppi_map('reflectivity', sweep=1, projection=ccrs.PlateCarree(),
                     lat_lines=lal, lon_lines=lol,
                     resolution='10m')
display.plot_range_ring(145, npts=360,
                        line_style='k-')
plt.show()
```

```python
# Plotting reflectivity using pyart on a cartopy map.
min_lon=110
max_lon=145
min_lat=-24
max_lat=2
lal = np.arange(min_lat, max_lat, 2)
lol = np.arange(min_lon, max_lon, 2)

fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
# Put a background image on for nice sea rendering.
ax.stock_img()

display = pyart.graph.RadarMapDisplayCartopy(radar)
display.plot_ppi_map('reflectivity', sweep=1, ax=ax, min_lon=min_lon,
                     max_lon=max_lon, min_lat=min_lat, max_lat=max_lat,
                     lat_lines=lal, lon_lines=lol,
                     resolution='10m')
display.plot_range_ring(145, npts=360,
                        line_style='k-')
ax.add_feature(cfeature.LAKES, zorder=0.5)
ax.add_feature(cfeature.RIVERS, zorder=0.5)
ax.gridlines()
plt.show()
```
